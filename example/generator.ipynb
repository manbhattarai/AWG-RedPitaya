```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
addr_width = 16
dac_data_width = 14;
x=np.arange(2**addr_width)
```

# Sine

```python
y = np.sin(2*np.pi/(2*2**addr_width)*x) #make this a unit amplitude
#z= int(y)
z = np.floor((2**(dac_data_width-1)-1)*(1-y))
np.savetxt("awg_half_sine.csv",z,fmt='%d')
```

## Double sided lorentzian

```python
def lor(x,x0,a,g):
    return (a*g**2/4/((x-x0)**2+g**2/4))

x_plus = 2**(addr_width-1)-2**(addr_width-2)
x_minus = 2**(addr_width-1)+2**(addr_width-6)
a=1
g= 2**11
y = lor(x,x_plus,a,g)+lor(x,x_minus,-a,g)
#z= int(y)
z = np.floor((2**(dac_data_width-1)-1)*(1-y))
np.savetxt("awg_double_lor.csv",z,fmt='%d')
plt.plot(y)
plt.title("Actual value")
plt.figure()
plt.plot(z,'r')
plt.title("Encoded value")
```
